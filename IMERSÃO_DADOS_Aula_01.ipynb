##Desafio7: Melhorar os gráficos (Matplotlib.pyplot)

Nessa parte do desafio, me concentrei inicialmente em construir os gráficos das séries estudadas até o momento.

Primeiro identificamos a proporção dos tratamentos através de um gráfico de setores, mais conhecido como gráfico de pizza, muito adequado para poucos dados com diferenças significativas entre eles.
 

````python
# Gráfico dos tipos de tratamento
import matplotlib.pyplot as plt

# Carregando a quantidade de casos(val) para cada tipo de tratamento(tipo).
qtd = dados['tratamento'].value_counts().values
tipo = dados['tratamento'].value_counts().index

# Plotando um gráfico de setor
plt.pie(qtd, labels = list(tipo), explode = (0, 0.2), autopct = '%1.1f%%', 
        labeldistance = 1.1, wedgeprops= {"ec": "#000000", "lw": 0.5})
plt.title('Formas de tratamento')
plt.legend()

# Assegura que o gráfico de setor seja desenhado como círculo
plt.axis('equal') 

# Oculta as informações no topo do gráfico
plt.show()
````

Depois, apresentamos os tempos de execução dos tratamentos, a partir de um gráfico de barras, onde é possível notar um número ligeiramente acima de 8000 experimentos para as 24h. Sendo este tempo o que mais ocorreu os experimentos.

````python
# Gráfico do tempo do experimento
import matplotlib.pyplot as plt

# Outra forma de carregar valores. Quantidades de cada tempo(qtd) e cada divisão de tempo (horas).
qtd = dados.tempo.value_counts()
horas = dados.tempo.unique()

# Discretizando a coordenada Y das horas.
plt.yticks(horas)

# Plotando um gráfico de barras horizontais
plt.barh(horas, qtd, height=10, align="center")
plt.title('Tempo dos experimentos do tratamento')

# Assimilando nomes para os eixos
plt.xlabel('Quantidade de experimentos')
plt.ylabel('Tempo (em horas)') 

# Oculta as informações no topo do gráfico
plt.show()
````

Por fim, apresentamos em um gráfico de rosca os valores da aplicação para das diferentes doses (D1 e D2).


````python
# Gráfico das doses
import matplotlib.pyplot as plt

# Carregando a quantidade de doses(qtd) para cada dose(dose).
qtd = dados.dose.value_counts()
dose = dados.dose.unique()

# Plotando um gráfico de setor
plt.pie(qtd, labels = list(dose), autopct = '%1.1f%%', pctdistance = 0.75, 
        wedgeprops= {"ec": "#000000", "lw": 0.5, "width": 0.5}, startangle= 90)
plt.title('Formas de tratamento\n')
plt.legend()

# Assegura que o gráfico de setor seja desenhado como círculo
plt.axis('equal') 

# Oculta as informações no topo do gráfico
plt.show()
````
